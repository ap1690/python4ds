```python

from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')

def clean(example_sent):
    stop_words = set(stopwords.words('english')) 
    word_tokens = tokenizer.tokenize(example_sent) 
    filtered_sentence = [w for w in word_tokens if not w.lower() in stop_words if len(w)>=1]
    return filtered_sentence
```

```python
cleaned_text=[clean(i) for i in content]
```

```python
initial=cleaned_text[0]
for i in range(1,len(cleaned_text)):
  initial+=cleaned_text[i]
```

```python
from collections import Counter
counts = Counter(initial)
print(counts)
```
